## Simple Chatbot

```python
import re
import random

# Define patterns and responses
pairs = [
    [
        r"hi|hello|hey",
        ["Hello!", "Hey there!", "Hi! How can I help you today?"]
    ],
    [
        r"how are you ?",
        ["I'm doing well, thank you!", "I'm good, thanks for asking."]
    ],
    [
        r"what's your name\??",
        ["I'm just a bot, but you can call me Harley. What's yours?", "I do have a name, but you can call me Harley. What about you?"]
    ],
    [
        r"how's the weather(?: outside)?\??",
        ["I'm not equipped to check the weather, but you can easily do so with a weather app!", "I'm not sure about the current weather. Is it sunny where you are?"]
    ],
    [
        r"where are you from\??",
        ["I'm from the internet! How about you?", "I'm just a virtual assistant, so I don't have a physical location. What about you?"]
    ],
    [
        r"what do you like to do for fun\??",
        ["As a bot, I enjoy assisting users and learning new things. How about you?", "I don't have hobbies, but I'm here to help you with any questions you have! What are your interests?"]
    ],
    [
        r"do you have any plans for today\??",
        ["I'm here to assist you, so no plans for me! What about you?", "My schedule is wide open. How about yours?"]
    ],
    [
        r"what's your favorite (movie|book|music)\??",
        ["I don't have personal preferences, but I can recommend some popular options! What are you interested in?", "As a bot, I don't have favorites, but I'm interested in learning about yours!"]
    ],
    [
        r"what's your favorite food\??",
        ["I don't eat, but I've heard people love pizza! What about you?", "I don't have taste buds, but I'm curious to know yours!"]
    ],
    [
        r"what are you doing this weekend\??",
        ["I don't have weekends, but I'm here to help you anytime! Any exciting plans for yours?", "Weekends are the same for me, assisting users like you! What about you?"]
    ],
    [
        r"what do you think about (politics|technology|sports)\??",
        ["As a bot, I don't have opinions, but I can provide information on those topics! What's your take?", "I'm neutral on such matters, but I'm here to discuss and answer any questions you have!"]
    ],
    [
        r"how's your day going\??",
        ["Every day is a good day when I get to assist users like you! How about yours?", "My day is going well, thanks for asking! How about yours?"]
    ],
    [
        r"what's your favorite place to visit\??",
        ["I don't travel, but I've heard the internet is a fascinating place! What about you?", "I'm confined to the digital realm, but I'd love to hear about your favorite destinations!"]
    ],
    [
        r"do you have any pets\??",
        ["I don't have pets, but I'm here to assist you with any questions you have! Do you have any pets?", "I'm a virtual assistant, so no pets for me. How about you?"]
    ],
    [
        r"quit",
        ["Bye! Take care.", "Goodbye! Have a great day."]
    ],
]

# Define a function to respond to user input
def chat():
    print("Hi! I'm Harley, a casual chatbot. How can I assist you today?")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print(random.choice(pairs[-1][1]))
            break
        else:
            response = generate_response(user_input)
            print("Harley:", response)

# Define a function to generate responses based on user input
def generate_response(user_input):
    for pattern, responses in pairs:
        if re.search(pattern, user_input, re.IGNORECASE):
            return random.choice(responses)
    return "I'm sorry, I didn't understand that."

# Start the conversation
chat()
```

## Summary:
This project is a Python-based chatbot named "Harley" that engages users in casual conversations on various topics using predefined patterns and random responses. It covers topics such as greetings, personal information, preferences, and general inquiries.
